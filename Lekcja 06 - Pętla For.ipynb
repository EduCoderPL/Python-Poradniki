### To ten... proponuję poznać pętlę :)

Pętla to taki Twór, w którym możemy powiedzieć, że dany kod możemy powtórzyć wiele razy.

```python
# Składnia najprostszej pętli - powtórzy dwie linijki kodu 5 razy.

for i in range(5):
    print("Arek")
    print("Mazurek")

print("Ta instrukcja już się nie powtórzy.")
```

```python
# Wykonuję ćwiczenie z poprzedniej części;

for i in range(30):
    print("Arek")
```

```python
# Wykonuję kolejne ćwiczenie z poprzedniej części

for i in range(120):
    print("Arek")
```

```python
# Popatrz, że teraz dużo łatwiej ogarnąć to ćwiczenie, na którym mogło nam parować pod czaszką :D
# Żeby kod był bardziej przejrzysty, niektóre z liczb umieściłem w zmiennych
# Popatrz, że nikt nam nie każe obliczać, ile dokładnie razy pętla się odpali. Dlatego możemy liczbę powtórzeń zapisać jako wyrażenie matematyczne. 
# Komputer sobie z tym poradzi :)

rozmiar_buta = 44
wiek_matki = 49

liczba_powtorek = 1137 * rozmiar_buta - wiek_matki + 12700

for i in range(liczba_powtorek):
    print("Arek")
```

```python
# Tyle razy odpalił się program

print(liczba_powtorek)
```

```python
# Czym dokładnie jest "i" w naszej pętli for?
# i to jest zmienna, która w każdej iiteracji (powtórzeniu) pętli wybiera wartości od 1 do 19 (nie 20, zauważ);
# znaczek ** to działanie potęgowania. 3 ** 5 to 3 * 3 * 3 * 3 * 3.   7 ** 3 = 7 * 7 * 7

for i in range(20):
    print(i, "do potęgi drugiej jest równe" + i ** 2)
```

```python
# Funkcję Range można odpalić też z dwoma parametrami (a nawet trzema)

for i in range(10, 20):
    print(i, i ** 2)

print()

for i in range(10, 20, 3):
    print(i, i ** 2)
```

```python
# Zadanie: Stworzyć listę napisów, a następnie dla każdego elementu zwrócić długośc napisu.

przedmioty = ("Polski", "Matematyka", "Przyroda", "Informatyka", "Niemiecki", "a", "abc")

for przedmiot in przedmioty:

  # Dodatkowy program, który pozwoli nam wybrać odpowiednią końcówkę wyrazu "litera"
    koncowka = ""
    if len(przedmiot) == 1:
        koncowka = "ę"
    elif len(przedmiot) <=4:
        koncowka = "y"
    else:
        koncowka = ""
        
    print(przedmiot + " ma "+ str(len(przedmiot))+ " liter"+koncowka)
```

```python
# Jak skrócić kod rysujący kwadrat?
# Przed skróceniem:
import turtle
turtle.reset()
turtle.speed(0)

turtle.forward(100)
turtle.left(90)
turtle.forward(100)
turtle.left(90)
turtle.forward(100)
turtle.left(90)
turtle.forward(100)
turtle.left(90)
```

```python
# Jak skrócić kod rysujący kwadrat?
# Po skróceniu:

import turtle
turtle.reset()
turtle.speed(0)
for i in range(4):
    turtle.forward(100)
    turtle.left(90)
```

```python
# Kto nam zabroni robić pętlę w pętli?

import turtle
turtle.reset()
turtle.speed(0)
for i in range(120):
    turtle.left(2);



    for i in range(4):
        turtle.forward(100)
        turtle.left(90)
```

```python
# A tutaj program, który prosi użytkownika o liczbę całkowitą (n), po czym rysuje wielokąt foremny o n bokach
# Wielokąt foremny: wielokąt, który ma wszystkie boki równe i kąty równe.
# Aby go narysować, należy pełny obrót (360 stopni) podzielić na n mniejszych obrotów (dlatego 360/n)

import turtle
turtle.reset()
turtle.speed(0)
n = int(input("Podaj n: "))
for i in range(150):
    turtle.left(2)
    for i in range(n):
        turtle.forward(800/n)
        turtle.left(360 / n)
```
