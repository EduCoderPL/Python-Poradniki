Wyrażenie, które zwraca prawdę lub fałsz:

```python
print(3<2)
print(10 >= 20)

# UWAGA!!!
# a == b oznacza porównanie dwóch zmiennych: a, b;
# a = b boznacza przypisanie zmiennej a wartości b;
print(20 == 10)
print(10 == 10)
```

### Działania logiczne:
**a or b** - zwraca prawdę, jeśli co najmniej jeden element jest prawdziwy
\

**a and b** - zwraca prawdę jeśli a i b są prawdziwe jednocześnie.

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)

print()

print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

# if

Przydatny, kiedy chcemy wykonać kod tylko przy spełnionym określonym warunku.
Zwróć uwagę, że:
- po słowie "if" ma pojawić sie warunek logiczny, czyli coś, co zwraca wartości True lub False;
- po warunku logicznym ma być dwukropek;
- następna linia po dwukropku powinna zaczynać się od wcięcia (najlepiej użyć klawisza TAB).
- to, że kod ma wcięcie decyduje o tym, że instrukcja ma być podległa warunkowi if.
- kod bez wcięcia już nie podchodzi pod if

```python
# Prosty przykład

wiek = int(input("Ile masz lat? "))

if wiek >= 18:
  print("Gratuluję, jesteś dorosły")
  print("Możesz wejść")
  print("Instrukcje z wcięciem wykonają się TYLKO jeśli warunek ifa będzie prawdziwy.")

print("To już nie należy do ifa, bo nie ma wcięcia.")

```

## else:
Jeśli chcemy, by oprócz zwykłego IF-a coś wykonało się w przypadku niespełnienia warunku, możemy użyć słowa kluczowego: else;

else nie potrzebuje warunku logicznego, ponieważ on się wykonuje, kiedy wcześniejszy warunek nie jest prawdziwy.



```python
# Prosty przykład

wiek = int(input("Ile masz lat? "))


if wiek >= 18:
  print("Gratuluję, jesteś dorosły")
else:
  print("Jesteś dzieckiem")
  print("Ta instrukcja wykona się tylko, jeśli warunek IF-a będzie nieprawdziwy.")


print("To już nie należy do ifa, bo nie ma wcięcia.")
```

## elif
Jeśli chcemy, by po nie spełnieniu warunku IF program sprawdził inny warunek, należy użyć słowa kluczowego elif oraz napisać nowy warunek logiczny.

elif-ów możemy wstawiać tyle, ile potrzebujemy.

```python
wiek = int(input("Ile masz lat? "))

if wiek >= 18:
  print("Gratuluję, jesteś dorosły")
elif wiek >=13:
  print("Jesteś nastolatkiem")
  print("To się wykona, jeśli pierwszy warunek będzie nieprawdziwy, ale drugi będzie prawdziwy.")
else:
  print("Jesteś dzieckiem")


print("To już nie należy do ifa, bo nie ma wcięcia.")
```

### Skrócony IF
Jeśli chcemy wykonać prosty mechanizm if-else do wybrania wartości zmiennej, możemy wykorzystać zapis skróconego ifa.

```python
# Wartość bezwzględna liczby
print("Program do podawania wartości bezwzględnej z liczby.")
liczba = int(input("Podaj dowolną liczbę: "))
liczba = liczba if liczba > 0 else -liczba

print(liczba)
```
